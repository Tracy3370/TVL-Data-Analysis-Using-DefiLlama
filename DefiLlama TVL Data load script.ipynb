```python
import requests
import json
import pandas as pd
```

```python
# make api request to get all protocols data from DefiLlama
response = requests.get("https://api.llama.fi/protocols")
print(response)
```

```python
# get the response in json formate
json_data = response.json()
```

```python
# Initialize the lists to hold data
ID = []
Name = []
symbol = []
description = []
chain = []
chains = []
Category = []
tvl = []
chainTvls = []
change_1h = []
change_1d = []
change_7d = []
```

```python
# loop through json data and parse data for interested users only
for data in json_data:
    if data['name'].lower() in User:
        ID.append(data['id'])
        Name.append(data['name'])
        symbol.append(data['symbol'])
        description.append(data['description'])
        chain.append(data['chain'])
        chains.append(data['chains'])
        Category.append(data['category'])
        tvl.append(data['tvl'])
        chainTvls.append(data['chainTvls'])
#         staking.append(data['staking'])
        change_1h.append(data['change_1h'])
        change_1d.append(data['change_1d'])
        change_7d.append(data['change_7d'])
#         fdv.append(data['fdv'])
#         mcap.append(data['mcap'])
```

```python
# Create pandas datafram for analysis and export to csv
Dict_data = {'ID' :ID,
            'Name' :Name,
            'Symbol' :symbol,
            'Description' :description,
            'Chain' : chain,
            'Chains':chains,
             'TVL':tvl,
             'Category' : Category,
             'ChainTVLs' :chainTvls,
             'Change_1h' :change_1h,
             'Change_1d' : change_1d,
             'Change_7d' : change_7d    }
df = pd.DataFrame(Dict_data )
```

```python
# Created Second Dataframe from dictonary of different chain tvls to create seperate columns
extracted_tvls = df['ChainTVLs'].apply(pd.Series)
```

```python
# Join Extracted Chain TVLS with original datafram horizontally. this will add additional columns
final = pd.concat([df, extracted_tvls], axis = 1)
```

```python
final.to_excel('Chainlink_user_TVL.xlsx', index=False)
```
